```python
import pandas as pd

# Load the dataset, parsing dates and setting the Month column as the index
df = pd.read_csv("datasets/worldwide_browser_trends.csv", index_col="Month", parse_dates=["Month"])

# Inspect the data to understand its structure
df.info()
df.head(10)
```

```python
# Calculate the six-month rolling average for each browser
rolling_six = df.rolling(window=6).mean()

# Display the first few rows of rolling_six to inspect
rolling_six.head(20)

# Plot the six-month rolling average
rolling_six.plot(title="6 Month Rolling Average for Browser Popularity")
```

```python
# Calculate the quarterly percentage change starting from 2009
# Each row represents monthly data, so we use periods=3 for quarterly changes
pct_change_quarterly = df.pct_change(periods=3) * 100  # Convert to percentage format
pct_change_quarterly = pct_change_quarterly.loc["2009":]  # Only include dates from 2009 onward

# Plot the quarterly percentage change for each browser
pct_change_quarterly.plot(subplots=True, figsize=(12,8), title="Quarterly Percentage Change in Browser Popularity")
```

```python
# Create an empty DataFrame and add Chrome's monthly data for each year as columns
chrome_trends = pd.DataFrame()
for year in ['2009', '2012', '2015', '2018']:
    chrome_trends_per_year = df.loc[year, ['Google Chrome']].reset_index(drop=True)
    chrome_trends_per_year.rename(columns={'Google Chrome': year}, inplace=True)
    chrome_trends = pd.concat([chrome_trends, chrome_trends_per_year], axis=1)

# Display the chrome_trends DataFrame
chrome_trends.head(12)

# Plot Chrome's performance for the selected years
chrome_trends.plot(title="Chrome Search Performance in 2009, 2012, 2015 & 2018")
```
